````python
from datasets import load_dataset
````

# Saving the Dataset

````python
ds = load_dataset("bigcode/pii-for-code", split="train")
````

````python
for i, datapoint in enumerate(ds):
    open(f".ipynb_checkpoints//ds_{i}.txt", "w").write(datapoint["content"])
````

# Run trufflehog on the files

````python
!trufflehog filesystem --directory .ipynb_checkpoints --no-verification --trace
````
